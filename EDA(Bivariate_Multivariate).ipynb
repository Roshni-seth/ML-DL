# Lineplot (numerical - numerical)

```python
import pandas as pd

# Assuming flights DataFrame is already loaded

# Check if 'month' column already contains numerical values
if pd.api.types.is_numeric_dtype(flights['month']):
    # If yes, convert to string and then to datetime objects
    flights['month'] = flights['month'].astype(str)
    flights['month'] = pd.to_datetime(flights['month'], format='%m').dt.month
else:
    # If not, proceed with original logic
    flights['month'] = pd.to_datetime(flights['month'], format='%b', errors='coerce').dt.month


# Now you can convert to integer if needed
flights['month'] = flights['month'].astype(int)

new=flights.groupby('year').sum().reset_index()
```

```python
sns.lineplot(x='year', y='passengers', data=new)
```

```python
sns.heatmap(flights.pivot_table(values='passengers',index='month',columns='year'))
```

```python
sns.clustermap((flights.pivot_table(values='passengers',index='month',columns='year')))
```
