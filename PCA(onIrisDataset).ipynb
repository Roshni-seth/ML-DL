```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
X=iris.data
```

```python
X_meaned=X-np.mean(X,axis=0)
```

```python
cov_matrix=np.cov(X_meaned,rowvar=False)
```

```python
eigenvalues,eigenvectors=np.linalg.eigh(cov_matrix)
```

```python
sorted_ind=np.argsort(eigenvalues)[::-1]
sorted_eigenvalue=eigenvalues[sorted_ind]
sorted_eigenvector=eigenvectors[:,sorted_ind]
```

```python
k=2
eigenvector_subset=sorted_eigenvector[:,0:k]
```

```python
X_reduced=np.dot(X_meaned,eigenvector_subset)
```

```python
principal_components_df = pd.DataFrame(data=X_reduced, columns=['PC1', 'PC2'])
principal_components_df.head()
```
